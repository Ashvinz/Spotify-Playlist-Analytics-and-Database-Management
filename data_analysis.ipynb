```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('spotify_playlist_data.csv')
data.head(2)
```

```python
data.isnull().sum()
```

```python
data.count()
```

```python
# ðŸŽµ 1. Popularity Distribution Plot (Histogram)

plt.figure(figsize=(8, 5))
plt.hist(data["Popularity"], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel("Popularity")
plt.ylabel("Number of Tracks")
plt.title(f"Popularity Distribution")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python


# ðŸ“Š 2. Track Duration vs. Popularity (Scatter Plot)

plt.figure(figsize=(8, 5))
plt.scatter(data["Duration (minutes)"], data["Popularity"], color='red', alpha=0.6, edgecolors='black')
plt.xlabel("Duration (minutes)")
plt.ylabel("Popularity")
plt.title(f"Duration vs. Popularity ")
plt.grid(linestyle='--', alpha=0.6)
plt.show()
```

```python

# ðŸ† 3. Top 5 Most Popular Tracks (Bar Chart)
top_tracks = data.nlargest(5, "Popularity")
plt.figure(figsize=(8, 5))
plt.barh(top_tracks["Track Name"], top_tracks["Popularity"], color='red', edgecolor='black', alpha=0.7)
plt.xlabel("Popularity")
plt.ylabel("Track Name")
plt.title(f"Top 5 Most Popular Tracks")
plt.gca().invert_yaxis() 
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```
